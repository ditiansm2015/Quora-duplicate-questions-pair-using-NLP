We will create custom features and add them to our dataframe.
The custom featurs are:


1.   **q1 length:** lenght of question 1
2.   **q2 length:** length of question 2
3.   **q1 words count:** number of words in question 1
4.   **q2 words count:** number of words in question 2
5.   **words common:** number of common unique words in question 1 and question 2
6.   **words total:** total number of words in question 1 + total number of words in question 2
7.   **words share:** words common/words share (words common divided by words share)

### **Adding custom features**

```python
# Custom feature 1 and 2 - lenght of question 1 and length of question 2

new_df['q1_len'] = new_df['question1'].str.len() 
new_df['q2_len'] = new_df['question2'].str.len()
```

```python
# We can see two new features added at the end 
new_df.head()
```

```python
# Custom feature 3 and 4 - number of words in question 1 and number of words in question 2

new_df['q1_num_words'] = new_df['question1'].apply(lambda row: len(row.split(" ")))
new_df['q2_num_words'] = new_df['question2'].apply(lambda row: len(row.split(" ")))
new_df.head()
```

```python
# Custom feature 5 - number of common unique words in question 1 and question 2

def common_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))   # it will first lower case the question1 column data then break the sentences into words and then since we are converting it into set, it will remove all the duplicates and each word will be occuring only once
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))   # it will first lower case the question2 column data then break the sentences into words and then since we are converting it into set, it will remove all the duplicates and each word will be occuring only once 
    return len(w1 & w2)     #  Applying intersection between w1 ad w2 which will give us the number (lenght) of common words between Ques 1 and Ques 2
```

```python
# Adding the new feature (words_common) which is the lenght of common words in a question pair (number of words common in ques1 and ques2 ) 

new_df['word_common'] = new_df.apply(common_words, axis=1)
new_df.head()
```

```python
# Custom feature 6- total number of unqiue words in question 1 + total number of unique words in question 2

def total_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))     # it will first lower case the question1 column data then break the sentences into words and then since we are converting it into set, it will remove all the duplicates and each word will be occuring only once
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))     # it will first lower case the question2 column data then break the sentences into words and then since we are converting it into set, it will remove all the duplicates and each word will be occuring only once
    return (len(w1) + len(w2))          # Applying union between w1 and w2 which will give us the number (lenght) of total unique words between Ques 1 and Ques 2
```

```python
new_df['word_total'] = new_df.apply(total_words, axis=1)
new_df.head()
```

```python
# Feature 7 - words common/words share (words common divided by words share)

new_df['word_share'] = round(new_df['word_common']/new_df['word_total'],2)
new_df.head()
```

```python
ques_df = new_df[['question1','question2']]
ques_df.head()
```

```python
final_df = new_df.drop(columns=['id','qid1','qid2','question1','question2'])
print(final_df.shape)
final_df.head()
```

```python
# Applying Bag of Words (BoW)

from sklearn.feature_extraction.text import CountVectorizer
# merge texts
questions = list(ques_df['question1']) + list(ques_df['question2'])   # Merging ques 1 column and ques 2 column and creating a new list (questions)

cv = CountVectorizer(max_features=3000)                               # we have decided to take only 3000 most occuring features and not all the features that we will get after applying BoW on questions list
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)   # applying BoW finally on questions list (which is a merger of ques 1 and ques 2) and then splitting the BoW features into two parts - first 3000 features in q1_arr and next 3000 features in q2_arr
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape

# Here you will get 6000 columns because 3000 features will be for BoW 1 (that we get by applying BoW on ques1) and another 3000 will be for BoW 2 (that we get by applying BoW on ques2)
```

```python
final_df = pd.concat([final_df, temp_df], axis=1)
print(final_df.shape)
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final_df.iloc[:,1:].values,final_df.iloc[:,0].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```

## **Advanced Features**

### 1. Token Features

cwc_min: This is the ratio of the number of common words to the length of the smaller question

cwc_max: This is the ratio of the number of common words to the length of the larger question

csc_min: This is the ratio of the number of common stop words to the smaller stop word count among the two questions

csc_max: This is the ratio of the number of common stop words to the larger stop word count among the two questions

ctc_min: This is the ratio of the number of common tokens to the smaller token count among the two questions

ctc_max: This is the ratio of the number of common tokens to the larger token count among the two questions

last_word_eq: 1 if the last word in the two questions is same, 0 otherwise

first_word_eq: 1 if the first word in the two questions is same, 0 otherwise




---


### 2. Length Based Features

mean_len: Mean of the length of the two questions (number of words)

abs_len_diff: Absolute difference between the length of the two questions (number of words)

longest_substr_ratio: Ratio of the length of the longest substring among the two questions to the length of the smaller question



---


### 3. Fuzzy Features

fuzz_ratio: fuzz_ratio score from fuzzywuzzy

fuzz_partial_ratio: fuzz_partial_ratio from fuzzywuzzy

token_sort_ratio: token_sort_ratio from fuzzywuzzy

token_set_ratio: token_set_ratio from fuzzywuzzy
